### Merging Adding and Concatenation

```python
import pandas as pd
import numpy as np
```

- Merging Two DataFrames

```python
employees = pd.DataFrame({
    'employee_id': [1,2,3,5,7],
    'name': ['peter', 'lambda', 'theta', 'pytho', 'rimi'],
    'dept': ['HR', 'IT', 'IT', "FINANCE", 'HR']
})
# DF2
salaries = pd.DataFrame({
    'employee_id': [1,2,6,8,7],
    'salary': [3200, 2100, 4500, 2300,7800],
    'bonus': [200, 400, 100, 300, 50]
})
 
```

```python
employees
```

```python
salaries
```

- merge both DF
- both can merge by 4 methods 
- 1st - inner, 2nd- outer, 3rd- left, 4th- right

```python
# on = kis column ke base par marge kare 
# how = kaise merge kare option
pd.merge(employees, salaries, on='employee_id', how='inner')
```

```python
# outer = union
pd.merge(employees, salaries, on='employee_id', how='outer')
```

```python
# left means employees df ke base par merge karo
pd.merge(employees, salaries, on='employee_id', how='left')
```

```python
# right means salaries df ke base par merge karo
pd.merge(employees, salaries, on='employee_id', how='right')
```

### concat

```python
data1 = pd.DataFrame({
    'A': ['A0','A1','A2'],
    'B': ['B0','B1','B2'],
    'C': ['C0','C1','C2']
})
data2 = pd.DataFrame({
    'A': ['D0','D1','D2'],
    'B': ['E0','E1','E2'],
    'C': ['F0','F1','F2']
})
```

```python
# By default use axis=0
pd.concat([data1,data2])
```

```python
pd.concat([data1,data2], axis=1)
```
