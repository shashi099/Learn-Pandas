### Missing Data Learn


```python
import numpy as np
import pandas as pd
```

```python
data = {
    "Name" : [ "Shashi", "Sonu", "Sumit", None],
    "Age" : [21, 10, None, None],
    "Salary" : [None, None, None, 1000]
}
df = pd.DataFrame(data)
df
```

- Check Missing Values


```python
# isna() provide us when data missing it-True And Data aviable is-False
# df.isnull()
df.isna()
```

- Count Each Column How many Data Missing

```python
df.isna().sum()
```

- Check kya ek bhi aisa row hai jisme null value nahi hai

```python
df.isna().any()
```

- Drop missing Data


```python
# Inside Bracket we can pass asix=0 | axis=1
df.dropna()
```

- Check atleast _kuch value hai jo Non Null hai a

```python
# Kam se kam 2 non null value hone chahiyea
df.dropna(thresh=2)
```

- Fill Missing Data with some Values

```python
# df.fillna(0)
values = ["Name": "Sahil", "Age": 100, "Salary": 2000]
df.fillna(value=values)
```
