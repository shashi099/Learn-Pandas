### GroupBY
- 🧠 What is GroupBy?
-- GroupBy means:
+ Group similar values together and perform calculations on them.

```python
import pandas as pd
data = pd.DataFrame({
    "Name":['A', 'B', 'C', 'D', 'E'],
    "Dept": ['IT', 'HR', 'IT', 'IT', 'HR'],
    "Salary": [2000, 5000, 1000, 1500, 2500]
})
data
```

Find the Average Salary of All Dept

```python
data.groupby("Dept")["Salary"].mean()
```

Find Total Salary of All Department

```python
data.groupby("Dept")["Salary"].sum()
```

find minimum salary of all Dept

```python
data.groupby("Dept")["Salary"].min()
```

Count Number of Employee in Each Dept

```python
data.groupby("Dept")["Salary"].count()
```

### Multiple Aggregations (Very Important 🔥🔥)

```python
data.groupby("Dept")["Salary"].agg(["mean","sum","max","count"])
```

### Group By Multiple Columns

```python
data.groupby(["Dept","Salary"]).count()
```

```python
# salary ka avg cal and index add ho jayega
data.groupby("Dept")["Salary"].mean().reset_index()
```

### PRACTICE QUESTION


```python
dataset = pd.DataFrame({
    "city" : ["Delhi", "Mumbai", "Delhi", "Mumbai", "Delhi"],
    "Sales" : [1000, 2000, 1500, 3000, 2000]
})
dataset
```

- 1️⃣ Find total sales per city

```python
dataset.groupby("city")["Sales"].sum()
```

- 2️⃣ Find average sales per city

```python
dataset.groupby("city")["Sales"].mean()
```

- 3️⃣ Find max sales per city


```python
dataset.groupby("city")["Sales"].max()
```
