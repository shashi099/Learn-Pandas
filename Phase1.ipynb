About Series in Pandas


```python
import numpy as np
import pandas as pd
```

```python
lebels = ['a', 'b', 'c']
my_list = [10, 20, 30]
my_dict = {'a': 21, 'b': 31, 'c': 41}
my_arr = np.array([1, 2, 3])
```

- In Pandas if we don't provide lebels then it automatically
  create lebels 0 1 2
- In Case of Dictonary lebels are these key
- And we can also provide lebels according to us

```python
# Automatically create lebels
pd.Series(my_list)
```

```python
# Automatically create lebels
pd.Series(my_arr)
```

```python
# Keys of Dict become lebels
pd.Series(my_dict)
```

```python
# we Can also provide lebels 
pd.Series(my_list, index=lebels)
```

### Data Frame
It's basically a 2-D Structured
(Rows + Column)
- can Create from Dictonary
- Also Can Create from list

- Created by Dict
- Then columns name will be dict 'Keys'

```python
info = {
    "Name": ["Shashi", "Rahul", "Monu"],
    "Age": [23, 22, 18],
    "Class" : ['XI', 'X', 'VII']
}
df = pd.DataFrame(info)
df
```

### Created By Lists
- Data Frame
- in which we get columns name as 0 1 2 ...

```python
product_data = [
    ['Laptop', 16000, 5],
    ['Phone', 12000, 8],
    ['Tablet', 10000, 3]
]
df2 = pd.DataFrame(product_data)
df2
```

```python
# Add Columns Name in Product
clm = ['Product', 'Price', 'Quantity']
df2 = pd.DataFrame(product_data, columns=clm)   
df2
```

- Selection of Columns

```python
# Select the column of Product
df2['Product']
```

- Select More than one columns

```python
df2[['Product', 'Price']]
```

- Add New Columns
Buy_Year = [2021, 2024, 2018]

```python
df2['Buy_Year'] = [2021, 2024, 2018]
df2
```

- Add A Column Total
- Total = Price * Quantity


```python
df2["Total"] = df2['Price'] * df2['Quantity']
df2
```

- Delete a Column

```python
df2.drop('Buy_Year', axis=1)  # Drop column Buy_Year
# df2.drop(0)  # Drop 1st row
```

- Selecting a row


```python
df2.loc[1]
```

Selection more than 1 row


```python
df2.loc[[0,1]]
```

- Selection Subsets of Rows And Column

```python
# Select first two rows price and quantity
df2.loc[[0,1], ['Price','Quantity']]
```

- Conditional Selection


```python
df2[df2.Price > 10000]
```
