About Series in Pandas


```python
import numpy as np
import pandas as pd
```

```python
lebels = ['a', 'b', 'c']
my_list = [10, 20, 30]
my_dict = {'a': 21, 'b': 31, 'c': 41}
my_arr = np.array([1, 2, 3])
```

- In Pandas if we don't provide lebels then it automatically
  create lebels 0 1 2
- In Case of Dictonary lebels are these key
- And we can also provide lebels according to us

```python
# Automatically create lebels
pd.Series(my_list)
```

```python
# Automatically create lebels
pd.Series(my_arr)
```

```python
# Keys of Dict become lebels
pd.Series(my_dict)
```

```python
# we Can also provide lebels 
pd.Series(my_list, index=lebels)
```
